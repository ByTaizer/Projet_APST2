## Q-Learning

```python
alpha = 0.5
gamma = 1
N = 3000
eps = 1
coeff = 0.01

q = np.zeros((16,4))
env = gym.make("FrozenLake-v1")
env.reset()
history = np.zeros(N)
for i in range(N):
    env.reset()
    terminated = False
    while not terminated:
        eps = np.exp(-coeff*i) #exponential decay
        state = env.s
        if np.random.uniform(0,1) < eps:
            action = np.random.choice(np.arange(4))
        else :
            if not np.any(q[state,:]):
                action = np.random.choice(np.arange(4))
            else:
                action = np.argmax(q[state,:])
        new_state, reward, terminated, truncated, info = env.step(action)
        history[i] += reward
        qmaxnew = np.max(q[new_state,:])
        q[state, action] = (1-alpha)*q[state, action] + alpha*(reward + gamma*qmaxnew)


env.close()
plt.plot(np.arange(N),history, c = 'b')
plt.xlabel('episode')
plt.ylabel('reward')
plt.title('evolution of reward per episode q-learning')
plt.show()
```

```python
env = gym.make("FrozenLake-v1", render_mode = 'human', is_slippery = True)
env.reset()

for i in range(4):
    env.reset()
    terminated = False
    while not terminated:
        state = env.s
        action = np.argmax(q[state,:])
        new_state, reward, terminated, truncated, info = env.step(action)

env.close()
```

```python
def performance(policy,n):
    env = gym.make("FrozenLake-v1")

    rewards = np.zeros(n)
    js = np.zeros(n)
    env.reset()
    for i in range(n):
        terminated = False
        env.reset()
        while not terminated:
            state = env.s
            action = np.argmax(policy[state,:])
            new_state, reward, terminated, truncated, info = env.step(action)
            js[i] += 1
            rewards[i] += reward
    env.close()
    return rewards, js
```

```python
n = 100
rewards, js = performance(q,n)
plt.subplot(1,2,1)
plt.scatter(np.arange(100),rewards)
plt.axhline(y = np.mean(rewards), c = 'r')
plt.xlabel('episode')
plt.ylabel('reward')
plt.subplot(1,2,2)
plt.scatter(np.arange(100),js)
plt.axhline(y = np.mean(js), c = 'r')
plt.xlabel('episode')
plt.ylabel('number of steps')
plt.show()
print("win rate :", np.mean(rewards))
```
